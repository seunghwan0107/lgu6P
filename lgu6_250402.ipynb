# loc vs iloc
-코드 비교

## loc

```python
#tips.loc[tips['day'] == 'Sat']
#문법 : tips.loc[행, 열]
tips.loc[0:1, ['total_bill', 'tip', 'day']]
```

## iloc

```python
tips.iloc[0:1, [0, 1, 4]]
```

```python
tips.loc[tips['day'] == 'Sat', ['total_bill', 'tip', 'day']].reset_index(drop=True)
```

```python
#tip.loc[행 조건식만 만들자!, :]
#tip.loc[ : , : ]
#totall_bill이 11이하인것만 조회
```

```python
tips.loc[tips['total_bill'] <= 11,:].reset_index()
```

```python
tips.loc[tips['time'] == 'Dinner',:].head(1)
```

```python
#time이 Dinner이면서 total_bill이 11이하인 것만 조회
result = tips.loc[tips['time'] == 'Dinner',:]
result.loc[result['total_bill'] <= 11, :].reset_index()
```

```python
tips.loc[(tips['time'] == 'Dinner') & (tips['total_bill'] <=11) , :].reset_index()
#조건식 2개
#tips.loc[(조건식1) & (조건식2), :]
```
